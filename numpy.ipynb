차원과 크기를 변경하는 reshape

```python
array1 = np.arange(10)
array2 = array1.reshape((2,5))
print(array1)
print(array1.ndim)
print(array2)
print(array2.shape)
print(array2.ndim)
array3 = array1.reshape((5,2))
print(array3)
# array4 = array1.reshape((4,3))
```

```python
array1 = np.arange(10)
array2 = array1.reshape((-1,5))
print(array1)
print(array1.ndim)
print(array2)
print(array2.shape)
print(array2.ndim)
array3 = array1.reshape((-1,2))
print(array3)
```

인덱싱과 슬라이싱  

```python
# 단일값 추출 
array = np.arange(start= 0,
    stop= 10)
print(array)
value = array[3]
print(value)
print(type(value))
```

```python
array1D = np.arange(start= 0,stop= 9)
print(array1D)
array2D = array1D.reshape((3,3))
print(array2D)
# 2차원[행,열]
print(array2D[0,0])
print(array2D[1,2])
print(array2D[2,1])
```

```python
# 슬라이싱
array = np.arange(start= 0,stop= 10)
print(array[0:2])
print(array[2:5])
print(array[:5])
print(array[5:])
print(array[:])
```

```python
# 팬시 인덱싱
array = np.arange(start= 0,stop= 10)
array2D = array.reshape((2,5))
print(array2D)
print(array2D[[0,1],2])
print(array2D[[0,1],0:2])
```

```python
# 불린 인덱싱
array = np.array([1,2,3,4,5,3,3])
array[array == 3] = 10
print(array)

# 리스트
list = [1,2,3,4,5,3,3]
list[list != 3] = 10
print(list)

# 조건문
array = np.arange(start= 0,stop= 10)
array>5
array[array > 5]s
```
